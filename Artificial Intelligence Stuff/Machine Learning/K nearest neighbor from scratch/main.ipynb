```python
import pandas as pd
import numpy as np
dftrf,dftsf=pd.read_csv('DataTrain_Tugas3_AI.csv',index_col=False,header=0),pd.read_csv('DataTest_Tugas3_AI.csv',index_col=False,header=0)
hurufka,hurufen=20,10 #max k and max n
mangatan=lambda x1,x2:np.abs(x1-x2)
def kaenen(kkk,dftr,dfts,test):
    j,akc,tokirim=0,0,[]
    for index2, risba2 in dfts.iterrows():
        klasjaraq,keleus=[],{}
        for index1, risba1 in dftr.iterrows():
            jaraq=sum(mangatan(risba2[i+1],risba1[i+1]) for i in range(5))
            klasjaraq.append({'keles':risba1[6],'jaraq':jaraq})
        klasjaraq.sort(key=lambda k:k['jaraq'])
        for i in range(kkk):
            if klasjaraq[i]['keles'] in keleus:
                keleus[klasjaraq[i]['keles']]+=1
            else:
                keleus[klasjaraq[i]['keles']]=1
        tokirim.append(int('%g'%(max(keleus,key=keleus.get))))
    if test==False: #validate if test false
        validator=list(dfts.iloc[:,6])
        xx=[x for x, y in zip(tokirim, validator) if x==y]
        akc=len(xx)/len(tokirim)
    return (akc,tokirim)
```

```python
fold,kkk=int(len(dftrf.iloc[:,6])/hurufen),[]
for i in range(hurufka):#start training with k-fold cross validation
    acc=0
    print('validating with k =',i+1)
    for j in range(hurufen):
        tes,tren=dftrf[j*fold:(j+1)*fold],dftrf.drop(dftrf.index[j*fold:((j+1)*fold)])
        akc,_=kaenen(i+1,tren,tes,False)
        print('fold',j+1,'accuracy =',akc)
        acc+=akc
    kkk.append(acc/hurufen)
    print('average:',acc/hurufen,'\nk =',i+1,'done')
[print('k =',idz+1,'| average =',var) for idz, var in enumerate(kkk)]
autok=max([(v,i) for i,v in enumerate(kkk)])[1]+1
print('biggest accuracy is when k =',autok)
```

```python
_,tokirim=kaenen(11,dftrf,dftsf,True) #testing with K = 11, or change the 11 with autok if you want an auto k generated
untukceesvi=pd.DataFrame(tokirim)
untukceesvi.to_csv("TebakanTugas3.csv",encoding='utf-8',index=False,header=False)
```
