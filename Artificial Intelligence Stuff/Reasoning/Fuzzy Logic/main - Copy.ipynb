```python
import pandas as pd
# import matplotlib as plt
import matplotlib.pyplot as plt
import numpy as np

def hl(l,r,ek):
    if ek<=l:
        return 0
    elif ek>r:
        return 1
    return (ek-l)/(r-l)
def ll(l,r,ek):
    if ek<=l:
        return 1
    elif ek>r:
        return 0
    return 1-((ek-l)/(r-l))


pend,hut=[],[]
df=pd.read_csv("DataTugas2.csv")
pend=df[' Pendapatan'].values
hut=df[' Hutang'].values
print(pend)
print(hut)
```
